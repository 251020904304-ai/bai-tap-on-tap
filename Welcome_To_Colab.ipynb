```
gpa = float(input("Nhập điểm trung bình (GPA): "))
if gpa >= 8.5:
 print("Xếp loại xuất xắc")
elif gpa >= 7.0:
 print("Xếp loại giỏi")
elif gpa >= 5.5:
 print("Xếp loại khá")
elif gpa >= 4.0:
 print("Xếp loại trung bình")
else:
 print("Xếp loại yếu")
```

```
n = int(input("Nhập số nguyên dương n: "))
tong = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        tong += i
print(f"Tổng các số chẵn từ 1 đến {n} là: {tong}")
```

```
mat_khau_dung = "python123"
so_lan_nhap = 0
toi_da = 3
dang_nhap_thanh_cong = False
while so_lan_nhap < toi_da:
    mat_khau_nhap = input(f"Mời nhập mật khẩu (Lần {so_lan_nhap + 1}): ")
    so_lan_nhap += 1
    if mat_khau_nhap == mat_khau_dung:
        print("Đăng nhập thành công")
        dang_nhap_thanh_cong = True
        break
    else:
        if so_lan_nhap < toi_da:
            print(f"Sai mật khẩu! Bạn còn {toi_da - so_lan_nhap} lần thử.")
if not dang_nhap_thanh_cong:
    print("Tài khoản bị khóa")
```
